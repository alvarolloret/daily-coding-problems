This problem was asked by Facebook.

Given a 32-bit integer, return the number with its bits reversed.

For example, given the binary number

1111 0000 1111 0000 1111 0000 1111 0000, return

  

0000 1111 0000 1111 0000 1111 0000 1111.

```python
number = 2131243
```

```python
def convert_int_to_32_bit(integer:int) -> str:
  return '{:032b}'.format(integer)

def validate_32_bit(binary_number:str) -> bool:
  return len(binary_number) == 32

def reverse_binary(binary_number:str) ->str:
  if validate_32_bit(binary_number):
    list_of_4_bit_number:list = [binary_number[i:i+4] for i in range(0, 32, 4)]
    reversed_list: list =  [list_of_4_bit_number[-i] for i in range(1, 9)]
    reversed_binary_string:str = "".join(reversed_list)
    return reversed_binary_string
  raise ValueError(f"The value {binary_number} is not a valid 32 bit ")

def convert_integer_to_binary_and_revert_its_bits(number:int)-> str:
  bits:str = convert_int_to_32_bit(number)
  reversed_bits: str = reverse_binary(bits)
  reversed_num  = int(reversed_bits, 32)
  print(f"Bits: {bits}")
  print(f"Reve: {reversed_bits}")
  print(f"Real: {reversed_num}")
  return reversed_num
```

```python
convert_integer_to_binary_and_revert_its_bits(number)
```

```python
NUM_BITS = 32

from IPython.core.debugger import Tracer

def reverse_bits(num):
    reversed_num = 0
    Tracer()()
    for i in range(NUM_BITS):
        jv = num >> i & 1
        reversed_num += jv << (NUM_BITS - i - 1)
    return reversed_num
```
