Good morning! Here's your coding interview problem for today.

This problem was recently asked by Google.

Given a list of numbers and a number k, return whether any two numbers from the list add up to k.

For example, given [10, 15, 3, 7] and k of 17, return true since 10 + 7 is 17.

Bonus: Can you do this in one pass?

## My solution

````python
def does_any_combination_of_the_list_of_numbers_add_up_to_k(list_of_numbers: list, value_k: int) -> bool:
  for index_i, i in enumerate(list_of_numbers):
    for index_j, j in enumerate(list_of_numbers):
      if index_i == index_j:
        break
      if i+j == value_k:
        return True
  return False

````

````python
NUMBERS = [10, 15, 3, 7]
value_k = 25
does_any_combination_of_the_list_of_numbers_add_up_to_k(NUMBERS, value_k)
````

Here's the solution to find whether any two numbers from the list add up to k:

```python
def two_sum(nums, k):
    seen = set()

    for num in nums:
        complement = k - num
        if complement in seen:
            return True
        seen.add(num)

    return False

# Example usage:
nums = [10, 15, 3, 7]
k = 17
result = two_sum(nums, k)
print(result)  # Output will be True since 10 + 7 is 17
```

This solution uses a set called `seen` to store the numbers as we iterate through the list. For each number, it calculates its complement (the number needed to reach k) and checks if that complement is already in the `seen` set. If it is, it means we've found two numbers that add up to k, so it returns `True`. If no pair is found after iterating through the entire list, it returns `False`.

This solution operates in one pass through the list, as it only requires a single iteration to check for pairs that add up to k.
