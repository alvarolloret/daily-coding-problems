### Solution using division:

1. Calculate the total product of all elements in the input array.
2. Iterate through the array and for each element at index `i`, divide the total product by the element at index `i` to get the product excluding that element.

Here is the Python code implementing the solution using division:

```python
def product_except_self(nums):
    total_product = 1
    for num in nums:
        total_product *= num

    result = [total_product // num for num in nums]
    return result

# Test cases
input1 = [1, 2, 3, 4, 5]
input2 = [3, 2, 1]

print(product_except_self(input1))  # Output: [120, 60, 40, 30, 24]
print(product_except_self(input2))  # Output: [2, 3, 6]
```

### Solution without using division:

Without using division, you can solve this problem by calculating the products of the left and right sides of each element separately, excluding the element itself. Then, multiply these left and right products to get the final result.

Here's the Python code implementing the solution without using division:

```python
def product_except_self(nums):
    length = len(nums)
    left_products = [1] * length
    right_products = [1] * length
    result = [1] * length

    # Calculate left products
    for i in range(1, length):
        left_products[i] = left_products[i - 1] * nums[i - 1]

    # Calculate right products
    for i in range(length - 2, -1, -1):
        right_products[i] = right_products[i + 1] * nums[i + 1]

    # Multiply left and right products to get the final result
    for i in range(length):
        result[i] = left_products[i] * right_products[i]

    return result

# Test cases
input1 = [1, 2, 3, 4, 5]
input2 = [3, 2, 1]

print(product_except_self(input1))  # Output: [120, 60, 40, 30, 24]
print(product_except_self(input2))  # Output: [2, 3, 6]
```

This solution has a time complexity of O(n), where n is the length of the input array, since it traverses the array multiple times to calculate the left and right products.

An XOR linked list is a more memory efficient doubly linked list. Instead of each node holding next and prev fields, it holds a field named both, which is an XOR of the next node and the previous node. Implement an XOR linked list; it has an add(element) which adds the element to the end, and a get(index) which returns the node at index.

If using a language that has no pointers (such as Python), you can assume you have access to get_pointer and dereference_pointer functions that converts between nodes and memory addresses.

````python
class Xor:
  listed = []
  def add(self, element):
      self.listed.append(element)
  def get(self, index):
      return self.listed[index]
````
